```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load CSV file
df = pd.read_csv("cars.csv")
df
```

## Check Duplicate Records

```python
duplicates = df.duplicated()
print("Duplicate Records:")
print(duplicates)
```

## Drop Duplicate Records

```python
print(df.duplicated().sum())
df_cleaned = df.drop_duplicates()
print(df_cleaned.shape)
print("Cleaned duplicate records")
```

### Check missing or null values

```python
missing_values = df.isnull().sum()
print("  Missing or Null Values:")
print(missing_values)
```

## Drop missing or null values

```python
df = df.dropna()
df.to_csv("cars_no_missing.csv", index=False)
print("Missing values removed & file saved as 'cars_no_missing.csv'")
```

## Show top 10 average MSRP (Price) per car brand (rounded figure)?

```python
top_10 = df.groupby("Make")['MSRP'].mean().sort_values(ascending=False).head(10).round(0).astype(int)
print(top_10)
```

## Show the average MSRP (price) per years (rounded figure)?

```python
df.groupby("Year")['MSRP'].mean().round(0).astype(int)
```

## Show average Engine HP (power) rounded figure and MPG (efficiency) rounded figure per year?

Hint: Average of (city mpg + highway MPG)

```python
df = pd.read_csv('cars.csv')
df.assign(mpg_combined=lambda x: (x['city mpg'] + x['highway MPG']) / 2).groupby('Year')[['Engine HP', 'mpg_combined']].mean().round(0).astype(int)
```

## What percentage of cars are Compact, Midsize, or Large?

```python
target_sizes = ['Compact', 'Midsize', 'Large']
size_percentage = (df['Vehicle Size'].value_counts(normalize=True).mul(100)[target_sizes])
print("Percentage of Cars by Size Category:")
print(size_percentage.round(2).astype(str) + '%')
```

## Which vehicle styles are the most sold (most common) in the dataset?

```python
print("Top 10 Most Common Vehicle Styles:")
df['Vehicle Style'].value_counts().head(10)
```

## Which car brands have the most vehicles priced at $40,000 or more and with a popularity score of at least 1000?

```python
brand_counts = df.query("MSRP >= 40000 & Popularity >= 1000")["Make"] \
.value_counts() \
.reset_index()
brand_counts.columns = ["Brand", "Vehicle Count"]
print(brand_counts)
print("\nTop 10 brands:")
print(brand_counts.head(10))
```
